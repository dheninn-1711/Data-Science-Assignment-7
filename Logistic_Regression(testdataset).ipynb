```python
#import library
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder
# Initialize LabelEncoder
label_encoder = LabelEncoder()
# Encode 'Sex' and 'Embarked' in data
data['Sex'] = label_encoder.fit_transform(data['Sex'])
data['Embarked'] = label_encoder.fit_transform(data['Embarked'])
data
```

```python
#drop unnecessary columns
data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True)
data
```

```python
#import libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#Handle missing values of X
X=data.iloc[:,1:]
# Handle missing values in X
Y=data.iloc[:,0]
X = X.apply(pd.to_numeric, errors='coerce')  # Ensure all data in X is numeric
# Split data into training and validation sets
X_train, X_valid, Y_train, Y_valid = train_test_split(X, Y, test_size=0.3, random_state=0)
# Initialize and train the model
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
#import library
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize
```

```python
# Predict on the validation set
Y_pred = model.predict(X_valid)
Y_prob = model.predict_proba(X_valid)
# Calculate metrics
accuracy = accuracy_score(Y_valid, Y_pred)
precision = precision_score(Y_valid, Y_pred, average='macro')  # Using 'micro', 'macro', or 'weighted'
recall = recall_score(Y_valid, Y_pred, average='macro')  
f1 = f1_score(Y_valid, Y_pred, average='macro')
Y_valid_bin = label_binarize(Y_valid, classes=[0, 1, 2])  # Modify classes based on your data
roc_auc = roc_auc_score(Y_valid, Y_prob, multi_class='ovr', average='macro')
#Print the values
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print(f'ROC AUC Score: {roc_auc}')
```

```python
# Plot ROC curve for each class
fpr, tpr, _ = roc_curve(Y_valid_bin[:, 1], model.predict_proba(X_valid)[:, 1])
plt.plot(fpr, tpr, color='red', label='Logistic Regression')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
# Coefficients of the logistic regression model
coefficients = model.coef_
feature_names = X.columns
#print values
for feature, coef in zip(feature_names, coefficients):
    print(f'Feature: {feature}, Coefficient: {coef}')
```

Significance:

Feature Coefficients: Positive coefficients increase the likelihood of survival, while negative coefficients decrease it. Larger magnitudes indicate stronger effects.


Interview Questions:

What is the difference between precision and recall?
Precision is the ratio of true positive predictions to the total number of positive predictions (true positives + false positives). It measures the accuracy of the positive predictions.

Recall (or Sensitivity) is the ratio of true positive predictions to the total number of actual positives (true positives + false negatives). It measures the ability of the model to identify all positive instances.

Cross-validation is a technique used to evaluate the performance of a model by partitioning the data into subsets.

Importance:

Helps in assessing how well the model generalizes to an independent dataset.

Offers a more reliable estimate of model performance by averaging results across different folds.

Ensures that each data point is used for both training and testing, leading to a more robust evaluation.
